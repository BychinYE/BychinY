## Problem 1

Let $f(x) = 2\sin x - \cos{2x}$. Write a function `beauty_derivative_plot` that takes a point $x$ and plots the function $f(x)$ and its derivative $f'(x)$ in the range $x \in [-2\pi, 2\pi]$. The function should also plot the tangent line to the function at the point $x$.

Example of output for $x = -3, -2, 1.7$:

```python
beauty_derivative_plot(-3)
```
<img src="https://i.ibb.co/W4pyPfTX/tangent-line-3-0.png" width="600"/>

```python
beauty_derivative_plot(-2)
```
<img src="https://i.ibb.co/F4zDq8m0/tangent-line-2-0.png" width="600"/>

```python
beauty_derivative_plot(1.7)
```
<img src="https://i.ibb.co/TBmLXg2h/tangent-line-1-7.png" width="600"/>

````python
import matplotlib.pyplot as plt
import numpy as np
def f(x):
    return 2 * np.sin(x) - np.cos(2 * x)#Определили функцию f(x)
def df(x):
    return 2 * np.cos(x) + 2 * np.sin(2 * x)#Определили производную функции f(x)
def beauty_derivative_plot(p):
    x = np.linspace(-2*np.pi, 2*np.pi, 1000)
    y = f(x)
    dy = df(x)
    fp = f(p)
    dfp = df(p)
    tangent_x = np.linspace(p - 1, p + 1, 100)
    tangent_y = fp + dfp * (tangent_x - p)#Уравнение касательной в точке p
    plt.figure(figsize=(10, 6))
    plt.plot(x, y, 'b', label=r'$f(x) = 2\sin x - \cos 2x$',color='blue', linewidth = 1)
    plt.plot(x, dy, 'r--', label=r"$f'(x) = 2\cos x + 2\sin 2x$", color = 'red', linewidth = 1, linestyle='--')
    plt.plot(tangent_x, tangent_y, 'g', linewidth=2, label=f'Tangent at x = {p:.2f}')
    plt.scatter([p], [fp], color='k', zorder=3)
    plt.vlines(p, 0, fp, linestyle='dashed', color='purple')
    plt.hlines(fp, 0, p, linestyle='dashed', color='purple')
    plt.annotate(f"f'({p:.2f}) = {dfp:.2f}",
             xy=(p, fp),
             xytext=(p - 0.5, fp - 1),
             arrowprops=dict(facecolor='black', arrowstyle='->', connectionstyle="arc3,rad=0.2"),
             fontsize=10, ha='center')
    #Подпись, заголовок и легенда
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Function, Derivative, and Tangent Line')
    plt.legend()
    plt.grid(True)
    plt.show()
#Пример работы в различных точках
beauty_derivative_plot(3)
beauty_derivative_plot(-2)
beauty_derivative_plot(1.7)
````

## Problem 2

The data in [populations.txt](https://drive.google.com/file/d/1Bq7qVzbZIOd058VobGE6iYjUcMFkbZMt/view?usp=drive_link) describes the populations of hares, lynxes and carrots in northern Canada during 20 years. Do the following exercises **without for-loops**.

- Plot the population size of each species for each year on a single graph. Add legend and axis labels to the plot.

````python
from google. colab import files
uploaded=files.upload()
data = np.loadtxt("populations.txt", skiprows=1)
years = data[:, 0]
hares = data[:, 1]
lynxes = data[:, 2]
carrots = data[:, 3]
plt.plot(years, hares, label="Зайцы")
plt.plot(years, lynxes, label="Рыси")
plt.plot(years, carrots, label="Морковь")
plt.xticks(np.arange(min(years), max(years) + 1, 2), rotation=45)
plt.xlabel("Year")
plt.ylabel("Population Size")
plt.legend()
plt.title("Population Sizes Of Hares, Lynxes, and Carrots Over Time")
plt.grid(True)
plt.show()
````

- Find which year each species had the largest population. Mark these years in the plot of populations.

````python
max_hare_year = years[np.argmax(hares)]
max_lynx_year = years[np.argmax(lynxes)]
max_carrot_year = years[np.argmax(carrots)]
plt.plot(years, hares, label="Зайцы")#точки на графике
plt.plot(years, lynxes, label="Рыси")
plt.plot(years, carrots, label="Морковь")
plt.xticks(np.arange(min(years), max(years) + 1, 2), rotation=45)
plt.scatter(max_hare_year, np.max(hares), color='blue', zorder=5)
plt.scatter(max_lynx_year, np.max(lynxes), color='orange', zorder=5)
plt.scatter(max_carrot_year, np.max(carrots), color='green', zorder=5)
plt.axvline(x=max_hare_year, color='blue', linestyle='--', alpha=0.5)#Перпендикуляр для зайцев
plt.axvline(x=max_lynx_year, color='orange', linestyle='--', alpha=0.5)#Перпендикуляр для рысей
plt.axvline(x=max_carrot_year, color='green', linestyle='--', alpha=0.5)#Перпендикуляр для моркови
plt.annotate(f"Макс. зайцы: {int(np.max(hares))} в {int(max_hare_year)}", (max_hare_year, np.max(hares)), textcoords="offset points", xytext=(0,3), ha='left')#легенда и подписи
plt.annotate(f"Макс. рыси: {int(np.max(lynxes))} в {int(max_lynx_year)}", (max_lynx_year, np.max(lynxes)), textcoords="offset points", xytext=(0,3), ha='center')
plt.annotate(f"Макс. морковь: {int(np.max(carrots))} в {int(max_carrot_year)}", (max_carrot_year, np.max(carrots)), textcoords="offset points", xytext=(0,3), ha='left')
plt.xlabel("Year")
plt.ylabel("Population Size")
plt.legend()
plt.title("Population Sizes Of Hares, Lynxes, and Carrots Over Time")
plt.grid(True)
plt.show()
plt.show()
````

- Assuming total population of hares and lynxes is 100%, plot stacked bar graph showing ratio of each specis for each year in the period.

- Find which species (hares or lynxes) has the largest population for each year. Print the result as [H, H, L, H, ...]. Plot a pie chart showing the ratio of "H" and "L" values.

- Plot the change in the hare population and in the lynx population for each year. Find the correlation coefficient between the both time series.

````python
hares_change = np.diff(hares)
lynxes_change = np.diff(lynxes)
plt.figure(figsize=(9, 4))
plt.plot(years[1:], hares_change, label='Hares Change',color='red')
plt.plot(years[1:], lynxes_change, label='Lynxes Change',color='green')
plt.xlabel('Year')
plt.ylabel('Population Change')
plt.title('Change in Population of Hares and Lynxes')
plt.legend()
plt.grid(True)
plt.show()
correlation = np.corrcoef(hares_change, lynxes_change)[0, 1]
print('Correlation coefficient =', correlation)
````

- Assume the population of hares in 1920 is unknown. Suggest a way to estimate this value. Compare an estimated value
with the true value and print a ratio of the error to the true value.

````python
years_without_1920 = years[years != 1920]
hares_without_1920 = hares[years != 1920]
estimated_hare_1920 = np.interp(1920, years_without_1920, hares_without_1920)#Линейная интерполяция
true_hare_1920 = hares[years == 1920][0]
error_ratio = np.abs(estimated_hare_1920 - true_hare_1920) / true_hare_1920
#Выводы результатов
print(f"~популяция зайцев в 1920: {estimated_hare_1920:.0f}")
print(f"популяция зайцев в действительности: {true_hare_1920:.0f}")
print(f"отношение ошибки к истинному значению: {error_ratio:.2f}")
````
